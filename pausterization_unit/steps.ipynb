```python
import numpy as np
```

```python
u_min = np.array([[30.0, 20.0, 0.0]])
u_max = np.array([[100.0, 100.0, 50.0]])
```

```python
u_start = np.array([[50.0, 50.0, 20.0]])
```

```python
steps = 11

random_steps = np.array([np.random.uniform(u_min, u_max)[0] for _ in range(steps)])
```

```python
random_steps
```

```python
u_prev = u_start
final_values = random_steps.copy()
for i in range(steps):
    final_values[i] = random_steps[i] - u_prev
    u_prev = random_steps[i]

final_values


```
